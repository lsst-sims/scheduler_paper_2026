```python
# Let's load up the DDFs vs time
from rubin_sim.data import get_data_dir
import os

ddf_file = os.path.join(get_data_dir(), 'scheduler/ts_ddf_array_b3abfb1.npz')
```

```python
ddf_obs = np.load(ddf_file)
obs_array = ddf_obs['obs_array'].copy()
ddf_obs.close()
```

```python
fig, ax = plt.subplots()
for ddf_name in np.unique(obs_array["target_name"]):
    if 'edfs' not in ddf_name:
        label = ddf_name.replace("ddf_", "").upper().replace("_A", "_a").replace("_B", "_b")
        indx = np.where(obs_array["target_name"] == ddf_name)[0]
        n_x = np.arange(indx.size) + 1
        ax.plot(np.sort(obs_array["mjd"][indx]), n_x, label=label, alpha=0.75)

label = "EDFS a+b"
indx = ['edfs' in note for note in obs_array["target_name"]]
n_x = np.cumsum(indx)
ax.plot(np.sort(obs_array["mjd"][indx]), n_x[indx], label=label, alpha=0.75)

ax.legend()
ax.set_xlabel("MJD")
ax.set_ylabel("N visits (all filters)")
ax.set_title("Pre-Scheduled DDF Observations")
fig.savefig("ddf_presched.pdf")
```
