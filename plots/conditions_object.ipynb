```python
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
%matplotlib inline
from rubin_scheduler.scheduler.model_observatory import ModelObservatory
from astropy.time import Time
```

```python
mo = ModelObservatory()
```

```python
conditions = mo.return_conditions()
```

```python
rot = (conditions.lmst*180/12, conditions.site.latitude)
hp.mollview(conditions.slewtime, unit='seconds', title='Slewtime', min=5, max=156, rot=rot)
plt.savefig('cond_slewtime.pdf')
```

```python
hp.mollview(conditions.skybrightness['r'], unit='$r$ mag arcsec$^{-2}$', title='Sky Brightness', rot=rot)
plt.savefig('cond_skybright.pdf')
```

```python
hp.mollview(conditions.airmass, unit='(unitless)', title='Airmass', max=3, min=1, rot=rot)
plt.savefig('cond_airmass.pdf')
```

```python
hp.mollview(conditions.fwhm_eff['r'], unit='arcsec', title='Effective Seeing, $r$',  max = 4, rot=rot)
plt.savefig('cond_seeing.pdf')
```

```python
hp.mollview(np.degrees(conditions.solar_elongation), unit='degrees', title='Solar Elongation', max=180, min=0, rot=rot)
plt.savefig('cond_selong.pdf')
```
